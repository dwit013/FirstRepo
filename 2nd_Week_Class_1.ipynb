```
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
```

```
from keras.models import Sequential
from keras.utils.np_utils import to_categorical
from keras.layers import Dense
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
df2 = pd.read_csv(io.BytesIO(uploaded['Admission_Predict.csv']))
```

```
df2
```

```
from sklearn.preprocessing import StandardScaler
```

```
y = df2["Chance of Admit "]
```

```
not_needed = ["Serial No.","Chance of Admit "]
X =  df2.drop(not_needed, axis=1)
```

```
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```
X.shape
```

```
model = Sequential()
model.add(Dense(units=10,input_dim=7,activation="relu"))
model.add(Dense(units=10,activation="relu"))
model.add(Dense(units=5,activation="relu"))
model.add(Dense(units=5,activation="relu"))

model.add(Dense(units=1,activation="sigmoid"))
model.compile(loss="mae",optimizer="sgd")
model.summary()
```

```
model.fit(X,y,epochs=150, validation_split=0.1)
```
